## Parsing all prograsm in page 1

```python
LAST_PAGE = 29

programs_dict ={
    'Title':[],
    'University':[],
    'Location':[],
    #'Mode':[],
    #'Language':[],
    'Start':[]  
}


```

```python
for page in range(1,LAST_PAGE+1):
    url = f"https://www.masterstudies.com/Political-Science/?page={page}"
    response = requests.get(url)
    programs_soup = BeautifulSoup(response.content, 'html.parser')
    programs_html = programs_soup.find_all(class_ = 'program-listitem')

    for program in programs_html:
        program_title = program.find('div', class_ = 'title').h4.text
        program_uni = program.find('div', class_ = 'school').text
        program_loc = program.find('span', class_ = 'location').text.strip().strip('                                                        +1 More')
        #program_mode = program.find('div', class_ = 'based').text.strip()
        #program_lang = program.find('div', class_ = 'languages').text.strip()
        program_start = 'None'
    #if (type(program.find('div', class_ = 'start').text.strip()) == None):
    #    programs_dict['Start'].append('N/A')
    #else:
     #   programs_dict['Start'].append(program.find('div', class_ = 'start').text.strip())
        programs_dict['Title'].append(program_title)
        programs_dict['University'].append(program_uni)
        programs_dict['Location'].append(program_loc)
        #programs_dict['Mode'].append(program_mode)
        #programs_dict['Language'].append(program_lang)
        programs_dict['Start'].append(program_start)
```

```python
programs_df = pd.DataFrame.from_dict(programs_dict)
```
